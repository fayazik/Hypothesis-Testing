**Question 1:** What is a null hypothesis (H₀) and why is it important in hypothesis testing?

**Answer:**

The Null Hypothesis (H₀) is a statement that there is no effect, no difference, or no relationship between variables. It represents the assumption that any observed difference is due to random chance.

Example:
If a company says the average delivery time is 10 minutes, the null hypothesis is:

H₀ : μ = 10

Importance of Null Hypothesis:

It provides a starting point for statistical testing.

It helps determine whether the sample evidence is strong enough to reject the claim.

It forms the basis for making statistical decisions.

**Question 2:** What does the significance level (α) represent in hypothesis testing?

**Answer:**

The significance level (α) is the probability of rejecting the null hypothesis when it is actually true.

In simple words, it is the risk of making a wrong decision.

Common values of α are:

0.05 (5%)

0.01 (1%)

Example:

If α = 0.05, it means there is a 5% chance of rejecting a true null hypothesis.

This error is called Type I Error.

**Question 3:** Differentiate between Type I and Type II errors.

**Answer:**

**Type I Error:**

Rejecting (H₀) when it is actually true.
-  Example: Concluding a new ad increases sales when it actually doesn’t.
- Controlled by significance level α (e.g., 0.05).

**Type II Error:**

Failing to reject (H₀) when it is false.
- Example: Missing a real improvement in customer satisfaction.
- Probability = β, and Power = 1 – β.

**Question 4:** Explain the difference between a one-tailed and two-tailed test. Give an example of each.

**Answer:**

**One-Tailed Test**

A one-tailed test checks if a parameter is greater than or less than a certain value.

Example:

H₀ : μ = 50
H₁ : μ > 50

This test checks only one direction.

**Two-Tailed Test**

A two-tailed test checks if a parameter is different from a certain value.

Example:

H₀ : μ = 50
H₁ : μ ≠ 50

This test checks both directions (greater or smaller).

**Question 5:** A company claims that the average time to resolve a customer complaint is 10 minutes.A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes. At α = 0.05, test the claim.

**Answer:**

Given:

    Claim: Average waiting time μ=10 minutes
    Sample size n = 9
    Sample mean x̄ = 12 minutes
    Sample Standard deviation s = 3 minutes
    Significance level α = 0.05

Step 1: Hypothesis

    H₀ : μ = 10
    H₁ : μ ≠ 10 (Two-tailed test)

Step 2: Test Statistic (t-score)

Since sample size < 30, use t-test

Formula:

    t= (x̄ - μ) / (s / √n)

Step 3: Calculation

    t = (12 − 10) / (3 / √9)
    t = 2 / (3 / 3)
    t = 2 / 1
    t = 2

Step 4: Critical Value

Degrees of freedom

    df = n − 1 = 9 − 1 = 8

    At α = 0.05 two-tailed test, t(critical) ≈ 2.306

Step 5: Decision

Since

    |t| = 2 < 2.306 , reject H₀

Step 6: Conclusion

There is not enough evidence to say the average complaint resolution time is different from 10 minutes.

So, the company's claim is acceptable.

**Question 6:** When should you use a Z-test instead of a t-test?

**Answer:**

**Use a Z-test when:**

- Sample size is large (n ≥ 30)
- Population standard deviation (σ) is known
- Data follows normal distribution

**Use a t-test when:**

- Sample size is small (n < 30)
- Population standard deviation is unknown
- Sample standard deviation is used instead.
